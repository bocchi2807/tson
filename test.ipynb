```python
#DATA MANIPULATION
df.item_price.dtype 
```

```python
df['item_price'] = df['item_price'].str.replace('$', '', regex=False).astype(float)
#dùng hàm apply và lambda để chuyển str về float
#lambda là 1 hàm mini siêu nhanh trên cùng một dòng mà không cần đặt tên cho nó (hàm ẩn danh)
#lambda dùng để thực hiện 1 phép tính or biến đổi đơn giản
```

```python
print(df['item_price'].dtype)
```

```python
df.head()
```

```python
#cách tạo thêm 1 colums mới
df['total_price']=df['quantity']*df['item_price']
```

```python
df.head()
```

```python
#how much was the revenue for the period in the dataset,
#tính tổng total price(1 cột)
revenue=df['total_price'].sum()
print(revenue)

```

```python
#which was the most-ordered items/
#item nào nhiều nhất(max)
#hàm group by
#hàm apply dùng để Là hàm dùng để áp dụng một quy tắc (hàm số) lên từng dòng hoặc từng cột của bảng dữ liệu.
x=df.groupby('item_name')['quantity'].sum()
x.sort_values(ascending=False).head(5)#từ lớn đến bé thì dùng ascending,còn k thì để trống
```
