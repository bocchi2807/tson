```python
# Task: Filter rows where Real Days > Scheduled Days
late_df=df.loc[df['Days for shipping (real)']>df['Days for shipment (scheduled)']]
print(f'number of delayed orders: {len(late_df)}')
print(late_df.shape)
late_df
```

```python
#Now, use NumPy to find exactly how many days late these orders were.
real=np.array(late_df['Days for shipping (real)'])
scheduled=np.array(late_df['Days for shipment (scheduled)'])
delays_day=real-scheduled
np.max(delays_day)
```

```python
late_df['Delay_Amount']=delays_day
#phải gán đúng giá trị vào bảng mình vừa mới khởi tạo,k gán lung tung
late_df
```

```python
#Is the company losing money on these late orders?
late_benefits=np.array(late_df['Benefit per order'])
np.mean(late_benefits)
```

```python
#gom nhóm(groupby)
#tính Tổng lợi nhuận và Tổng giá trị hàng của từng loại mặt hàng.
#groupby giúp tính tổng của mỗi thứ riêng biệt
category_analysis=df.groupby('Category Name')[['Benefit per order','Product Price']].sum()
print(category_analysis.head(5))
```

```python
#Profit Margin
#Từ bảng category_analysis vừa tạo:
total_benefits=np.array(category_analysis['Benefit per order'])
total_prices=np.array(category_analysis['Product Price'])
margins=total_benefits/total_prices
category_analysis['Profit_Margins']=margins
category_analysis.head()
```

```python
top_3_lai=category_analysis.sort_values(by='Profit_Margins',ascending=False).head(3)
top_3_lai
```
